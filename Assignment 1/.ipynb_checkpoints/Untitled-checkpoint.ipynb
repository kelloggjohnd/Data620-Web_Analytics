In this assignment, we are asked to create the nodes and edges for a basic network graph, such as the Krackhardt kite.

```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
```

We start off by making an empty graph to start putting things in

```python
w2 = nx.Graph()
```

With this shell, we can add our node information to the graph

```python
nodes = ['John', 'Tanya', 'Gabriel', 'Marie', 
        'Pietro', 'Pete', 'Ceil', 'Maddie', 'Sarah']
w2.add_nodes_from(nodes)
```

With the nodes in place, we can now add the edges or connections.

```python
edges = [('John', 'Tanya'),
         ('John', 'Gabriel'),
         ('John', 'Marie'),
         ('John', 'Pete'),
         ('Tanya', 'Marie'),
         ('Tanya', 'Garth'),
         ('Tanya', 'Pietro'),
         ('Gabriel', 'Pete'),
         ('Gabriel', 'Marie'),
         ('Pietro', 'Marie'),
         ('Pietro', 'Garth'),
         ('Marie', 'Pete'),
         ('Marie', 'Garth'),
         ('Pete','Garth'),
         ('Ceil', 'Pete'),
         ('Ceil', 'Garth'),
         ('Ceil', 'Maddie'),
         ('Maddie','Sarah' )
        ]
w2.add_edges_from(edges)
```

We can now create the graph

```python
nx.draw(w2, node_color='#A0CBE2',edge_color="purple",width=4,edge_cmap=plt.cm.Blues, node_size=1500, with_labels=True)
plt.show()
```

Going to try a different method to create the graph, this time using a position for each of the nodes.

```python
G=nx.Graph()

G=nx.krackhardt_kite_graph(G)
# Create the graph using the built in generator of Networkx

G.nodes()

position_dict ={'John':(-4, 1),
                'Tanya':(-4, -1),
                'Gabriel':(0, 2),
                'Marie':(0, 0),
                'Pietro':(0, -2),
                'Pete':(12, 0),
                'Ceil':(4, -1),
                'Maddie':(16, 0),
                'Sarah':(4, 1),
                'Michelle':(8, 0)}

G = nx.relabel_nodes(G, {0: 'John', 1: 'Tanya',  2: 'Gabriel', 3: 'Marie', 4: 'Pietro',
                         5: 'Pete',  6: 'Ceil', 7: 'Maddie', 8: 'Sarah', 9: 'Michelle'}, copy=False)

nx.draw(G, position_dict, node_color='#A0CBE2',edge_color="purple",width=4,edge_cmap=plt.cm.Blues, node_size=1500, with_labels=True)

plt.show()

```
